```python
# 1. Retrieve for each employee his/her first name, last name, and address.

df1 = pd.read_sql_query("SELECT FirstName, LastName, Address FROM employees;",conn)
df1.head()
```

```python
# 2. Retrieve for each customer his/her first name, last name, and invoices ID.

df = pd.read_sql_query("SELECT FirstName, LastName, InvoiceId FROM customers c, invoices i WHERE i.customerid = c.customerid;",conn)
df.head()
```

```python
# 3. Retrieve for each customer his/her first name, last name, and the sum for all his/her invoices amounts.

df3 = pd.read_sql_query("SELECT FirstName, LastName, SUM(TOTAL) FROM customers c, invoices i WHERE i.customerid = c.customerid GROUP BY i.customerid;",conn)
df3.head()
```

```python
# 4. Retrieve for each album, title, artist name, and genre name.

df4 = pd.read_sql_query("SELECT Albums.Title AS AlbumTitle, Artists.Name AS ArtistName, Genres.Name AS GenreName FROM Albums JOIN Artists ON Albums.ArtistId = Artists.ArtistId JOIN  Tracks ON Albums.AlbumId = Tracks.AlbumId JOIN  Genres ON Tracks.GenreId = Genres.GenreId GROUP BY Albums.Title, Artists.Name, Genres.Name;", conn)
df4.head() 
```

```python
# 5. Check if there are nulls values in part a, and replace it with zero, then show the dataframe head.


df  = pd.read_sql_query("SELECT FirstName, LastName, Address FROM employees;",conn)

if df.isnull().sum().any():
    df.fillna(0, inplace=True)
else:
    print("No null values found in the DataFrame.")

print("\nDataFrame after checking and replacing null values if necessary:\n")
print(df.head())
```

```python
# 6. Create a fact dataframe that includes all information regarding transactions of the customers. The dataframe should include **customer info, invoices’ items, and tracks with their genres

df = pd.read_sql_query("SELECT * FROM customers c, invoices i, invoice_items ii, tracks t, genres g WHERE i.customerid = c.customerid AND i.invoiceid = ii.invoiceid AND ii.trackid = t.trackid AND t.genreid = g.genreid;",conn)
df.head()
```
